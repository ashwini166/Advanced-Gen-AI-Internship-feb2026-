1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.

```python
# Stored credentials
correct_username = "admin"
correct_password = "1234"

# Taking input from the user
username = input("Enter username: ")
password = input("Enter password: ")

# Checking credentials
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer
Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

```python
# List of student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Analyze results
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print final result
print("Total Students:", len(marks))
print("Number of Pass Students:", pass_count)
print("Number of Fail Students:", fail_count)
```

3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]

```python
# Original list of names
names = [" Alice ", "bob", " CHARLIE "]

# Clean the names
cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()  # remove spaces and convert to lowercase
    cleaned_names.append(cleaned_name)

# Print the cleaned list
print("Cleaned Names:", cleaned_names)
```

4. Message Length Analyzer
Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("⚠️ This message is longer than 10 characters.")
```

Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

```python
# System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Count ERROR entries
error_count = logs.count("ERROR")

# Print the result
print("Total ERROR messages:", error_count)
```
