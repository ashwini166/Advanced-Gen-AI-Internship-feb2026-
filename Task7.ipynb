Problem 1: Smart Light Controller

Problem Statement-
Create a class to control a smart light that can be turned ON or OFF and display its status.

```python
# Smart Light Controller Class
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.is_on = False  # Default state is OFF

    def turn_on(self):
        self.is_on = True
        self.display_status()

    def turn_off(self):
        self.is_on = False
        self.display_status()

    def display_status(self):
        if self.is_on:
            print(f"{self.light_name} is ON")
        else:
            print(f"{self.light_name} is OFF")


# Main Program
if __name__ == "__main__":
    # Asking user input
    light_name = input("Light Name: ")
    action = input("Action: ")

    light = SmartLight(light_name)

    if action.upper() == "ON":
        light.turn_on()
    elif action.upper() == "OFF":
        light.turn_off()
    else:
        print("Invalid Action. Please enter ON or OFF.")
```

Problem 2: Employee ID Card System

Problem Statement-
Create a class that stores employee name, ID, and department and displays ID card details.




```python
# Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("\n----- Employee ID Card -----")
        print(f"Employee Name : {self.name}")
        print(f"Employee ID   : {self.emp_id}")
        print(f"Department    : {self.department}")
        print("----------------------------")


# Main Program
if __name__ == "__main__":
    # Taking user input
    name = input("Employee Name: ")
    emp_id = input("Employee ID: ")
    department = input("Department: ")

    # Creating Employee object
    employee = Employee(name, emp_id, department)

    # Displaying ID card details
    employee.display_id_card()
```

Problem 3: Mobile Contact Record

Problem Statement-
Create a class to store contact name and phone number and display contact information.


```python
# Mobile Contact Record System

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# Main Program
if __name__ == "__main__":
    # Taking user input
    name = input("Contact Name: ")
    phone = input("Phone Number: ")

    # Creating Contact object
    contact = Contact(name, phone)

    # Displaying contact details
    contact.display_contact()
```

Problem 6: Delivery Address Manager

Create a class that stores customer name and delivery address and prints delivery details.
